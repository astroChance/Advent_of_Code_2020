```python
"""
--- Day 4: Passport Processing ---
You arrive at the airport only to realize that you grabbed your North Pole 
Credentials instead of your passport. While these documents are extremely similar, 
North Pole Credentials aren't issued by a country and therefore aren't actually 
valid documentation for travel in most of the world.

It seems like you're not the only one having problems, though; a very long line 
has formed for the automatic passport scanners, and the delay could upset your 
travel itinerary.

Due to some questionable network security, you realize you might be able to solve 
both of these problems at the same time.

The automatic passport scanners are slow because they're having trouble detecting
which passports have all required fields. The expected fields are as follows:

byr (Birth Year)
iyr (Issue Year)
eyr (Expiration Year)
hgt (Height)
hcl (Hair Color)
ecl (Eye Color)
pid (Passport ID)
cid (Country ID)
Passport data is validated in batch files (your puzzle input). Each passport is 
represented as a sequence of key:value pairs separated by spaces or newlines. 
Passports are separated by blank lines.

Here is an example batch file containing four passports:

ecl:gry pid:860033327 eyr:2020 hcl:#fffffd
byr:1937 iyr:2017 cid:147 hgt:183cm

iyr:2013 ecl:amb cid:350 eyr:2023 pid:028048884
hcl:#cfa07d byr:1929

hcl:#ae17e1 iyr:2013
eyr:2024
ecl:brn pid:760753108 byr:1931
hgt:179cm

hcl:#cfa07d eyr:2025 pid:166559648
iyr:2011 ecl:brn hgt:59in

The first passport is valid - all eight fields are present. The second passport
is invalid - it is missing hgt (the Height field).

The third passport is interesting; the only missing field is cid, so it looks 
like data from North Pole Credentials, not a passport at all! Surely, nobody 
would mind if you made the system temporarily ignore missing cid fields. Treat 
this "passport" as valid.

The fourth passport is missing two fields, cid and byr. Missing cid is fine, 
but missing any other field is not, so this passport is invalid.

According to the above rules, your improved system would report 2 valid passports.

Count the number of valid passports - those that have all required fields. 
Treat cid as optional. In your batch file, how many passports are valid?
"""
```

```python
import re
```

```python
## puzzle input

inp_loc = './day_4_input.txt'


with open(inp_loc, 'r') as txtfile:
    content = txtfile.read()
    entries = content.split('\n\n')

print(len(entries))
print(entries[0])
```

```python
valid_test_case = "ecl:gry pid:860033327 eyr:2020 hcl:#fffffd\nbyr:1937 iyr:2017 cid:147 hgt:183cm"
invalid_test_case = "iyr:2013 ecl:amb cid:350\neyr:2023 pid:028048884\nhcl:#cfa07d byr:1929"
```

```python
def fields_list(entry):
    """
    Accepts a single passport entry,
    returns list of fields present
    """
    entry = re.split(' |\n', entry)
    fields = []
    for i in entry:
        fields.append(re.split(':',i)[0])
    return fields

def passport_field_check(entry):
    """
    Checks if passport contains all required
    (hard-coded) entry fields
    """
    required_fields = ['byr', 'iyr', 'eyr', 'hgt', 'hcl', 'ecl', 'pid']
    fields = fields_list(entry)
    for i in fields:
        i = i.lower()
    if set(required_fields).issubset(fields):
        return True
    else:
        return False
```

```python
print("Valid test case: {}".format(passport_field_check(valid_test_case)))
print("Valid test case: {}".format(passport_field_check(invalid_test_case)))
```

```python
for i in re.split(' |\n', valid_test_case):
    print(i)
```

```python
## Answer part 1

valid_count = 0
invalid_count = 0

for E in entries:
    if passport_field_check(E):
        valid_count += 1
    else:
        invalid_count += 1
        
print("Number of valid passports: {}".format(valid_count))
print("Number of invalid passports: {}".format(invalid_count))
```

```python
"""
--- Part Two ---
The line is moving more quickly now, but you overhear airport security 
talking about how passports with invalid data are getting through. 
Better add some data validation, quick!

You can continue to ignore the cid field, but each other field has 
strict rules about what values are valid for automatic validation:

byr (Birth Year) - four digits; at least 1920 and at most 2002.
iyr (Issue Year) - four digits; at least 2010 and at most 2020.
eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
hgt (Height) - a number followed by either cm or in:
If cm, the number must be at least 150 and at most 193.
If in, the number must be at least 59 and at most 76.
hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
pid (Passport ID) - a nine-digit number, including leading zeroes.
cid (Country ID) - ignored, missing or not.

Your job is to count the passports where all required fields are both 
present and valid according to the above rules. Here are some example values:

byr valid:   2002
byr invalid: 2003

hgt valid:   60in
hgt valid:   190cm
hgt invalid: 190in
hgt invalid: 190

hcl valid:   #123abc
hcl invalid: #123abz
hcl invalid: 123abc

ecl valid:   brn
ecl invalid: wat

pid valid:   000000001
pid invalid: 0123456789


Here are some invalid passports:

eyr:1972 cid:100
hcl:#18171d ecl:amb hgt:170 pid:186cm iyr:2018 byr:1926

iyr:2019
hcl:#602927 eyr:1967 hgt:170cm
ecl:grn pid:012533040 byr:1946

hcl:dab227 iyr:2012
ecl:brn hgt:182cm pid:021572410 eyr:2020 byr:1992 cid:277

hgt:59cm ecl:zzz
eyr:2038 hcl:74454a iyr:2023
pid:3556412378 byr:2007


Here are some valid passports:

pid:087499704 hgt:74in ecl:grn iyr:2012 eyr:2030 byr:1980
hcl:#623a2f

eyr:2029 ecl:blu cid:129 byr:1989
iyr:2014 pid:896056539 hcl:#a97842 hgt:165cm

hcl:#888785
hgt:164cm byr:2001 iyr:2015 cid:88
pid:545766238 ecl:hzl
eyr:2022

iyr:2010 hgt:158cm hcl:#b6652a ecl:blu byr:1944 eyr:2021 pid:093154719

Count the number of valid passports - those that have all required fields 
and valid values. Continue to treat cid as optional. In your batch file, 
how many passports are valid?
"""
```

```python
valid1 = 'pid:087499704 hgt:74in ecl:grn iyr:2012 eyr:2030 byr:1980\nhcl:#623a2f'
valid2 = 'eyr:2029 ecl:blu cid:129 byr:1989\niyr:2014 pid:896056539 hcl:#a97842 hgt:165cm'
valid3 = 'hcl:#888785\nhgt:164cm byr:2001 iyr:2015 cid:88\npid:545766238 ecl:hzl\neyr:2022'
valid4 = 'iyr:2010 hgt:158cm hcl:#b6652a ecl:blu byr:1944 eyr:2021 pid:093154719'

invalid1 = 'eyr:1972 cid:100\nhcl:#18171d ecl:amb hgt:170 pid:186cm iyr:2018 byr:1926'
invalid2 = 'iyr:2019\nhcl:#602927 eyr:1967 hgt:170cm\necl:grn pid:012533040 byr:1946'
invalid3 = 'hcl:dab227 iyr:2012\necl:brn hgt:182cm pid:021572410 eyr:2020 byr:1992 cid:277'
invalid4 = 'hgt:59cm ecl:zzz\neyr:2038 hcl:74454a iyr:2023\npid:3556412378 byr:2007'

valid_test_entries = [valid1, valid2, valid3, valid4]
invalid_test_entries = [invalid1, invalid2, invalid3, invalid4]
```

```python
def passport_check(entry):
    """
    Accepts single passport entry
    Checks for validity criteria (hard-coded)
    Returns True (valid) or False (invalid)
    """
    ###########
    # Criteria:
    year_digit_length = 4
    birth_range = range(1920, 2003) # end year is non-inclusive
    issue_range = range(2010, 2021)
    exp_range = range(2020, 2031)
    height_units = ['cm', 'in']
    height_range_cm = range(150, 194)  # currently only accepts integers
    height_range_in = range(59, 77)
    eye_colors = ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
    ###########
    
    required_fields = ['byr', 'iyr', 'eyr', 'hgt', 'hcl', 'ecl', 'pid']
    
    validity_list = []
    
    if passport_field_check(entry):
        for E in re.split(' |\n', entry):
            current_field = re.split(':',E)[0]
            try:
                current_value = re.split(':',E)[1]
            except IndexError:
                continue
            
            #Check birth year
            if current_field.lower() == required_fields[0]:
                if int(current_value) in birth_range:
                    validity_list.append(1)
                else:
                    validity_list.append(0)
            
            # check issue year
            if current_field.lower() == required_fields[1]:
                if int(current_value) in issue_range:
                    validity_list.append(1)
                else:
                    validity_list.append(0)
                    
            # check expiration year
            if current_field.lower() == required_fields[2]:
                if int(current_value) in exp_range:
                    validity_list.append(1)
                else:
                    validity_list.append(0)
                    
            # check height
            if current_field.lower() == required_fields[3]:
                H = (re.match(r"(\d+)\s*(.*)$", current_value))
                unit = H.group(2)
                height = int(H.group(1))
                
                if unit.lower() == height_units[0]:  # cm
                    if height in height_range_cm:
                        validity_list.append(1)
                    else:
                        validity_list.append(0)
                if unit.lower() == height_units[1]:  # in
                    if height in height_range_in:
                        validity_list.append(1)
                    else:
                        validity_list.append(0)
            
            # check hair color
            if current_field.lower() == required_fields[4]:
                if (len(current_value[1:]) == 6) and (current_value[1:].isalnum()):
                    validity_list.append(1)
                else:
                    validity_list.append(0)
            
            # check eye color
            if current_field.lower() == required_fields[5]:
                if current_value.lower() in eye_colors:
                    validity_list.append(1)
                else:
                    validity_list.append(0)
            
            # check passport ID
            if current_field.lower() == required_fields[6]:
                if len(current_value) == 9 and current_value.isdecimal():
                    validity_list.append(1)
                else:
                    validity_list.append(0)
                    
    if sum(validity_list) == 7:
        return True
    else:
        return False
```

```python
## check test cases

print("Valid tests:")
for i in valid_test_entries:
    print(passport_check(i))
    
print("\nInvalid tests:")
for i in invalid_test_entries:
    print(passport_check(i))
```

```python
## answer #2

valid_entries = 0
invalid_entries = 0

for i in entries:
    if(passport_check(i)):
        valid_entries += 1
    else:
        invalid_entries += 1
        
print("Number of valid passports: {}".format(valid_entries))
print("Number of invalid passports: {}".format(invalid_entries))
```
